# ***3. Data Wrangling***

```python
# Make a copy of the dataset to preserve original
df_clean = df.copy()

# ---- 1. Drop irrelevant or empty columns (if applicable) ----
# Example: drop 'comments' if not useful
if 'comments' in df_clean.columns:
    df_clean.drop('comments', axis=1, inplace=True)

# ---- 2. Fix age outliers ----
# Keep only realistic ages (e.g., 18–65)
df_clean = df_clean[(df_clean['Age'] >= 18) & (df_clean['Age'] <= 65)]

# ---- 3. Standardize gender values ----
def clean_gender(gender):
    gender = str(gender).strip().lower()
    if gender in ['male', 'm', 'man']:
        return 'Male'
    elif gender in ['female', 'f', 'woman', 'cis female']:
        return 'Female'
    elif 'trans' in gender:
        return 'Transgender'
    else:
        return 'Other'

df_clean['Gender'] = df_clean['Gender'].apply(clean_gender)

# ---- 4. Handle missing values ----
# You can drop or fill based on your needs; example below:
# Drop rows where key columns are missing
df_clean.dropna(subset=['Age', 'Gender', 'Country', 'treatment'], inplace=True)

# Optionally fill categorical missing values with 'Unknown'
df_clean.fillna('Unknown', inplace=True)

# ---- 5. Reset index after cleanup ----
df_clean.reset_index(drop=True, inplace=True)

# View cleaned data
df_clean.head()
```

# What all manipulations have you done and insights you found?


---

###  **Data Manipulations Performed:**

1. **Removed Irrelevant Data**

   * Dropped unnecessary columns like `comments`.

2. **Filtered Age**

   * Removed unrealistic age entries (kept age between 18 and 65).

3. **Standardized Gender**

   * Unified variations of gender entries (e.g., "male", "M", "Man" → "Male").

4. **Handled Missing Values**

   * Dropped rows with critical missing data (e.g., `treatment`, `country`).
   * Filled less important missing values with `'Unknown'`.

5. **Removed Duplicates**

   * Identified and removed duplicate records.

# 15. Age vs Treatment ( pair plot )

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant variables for pairplot
# Encode categorical 'treatment' column for coloring
df_pair = df_clean.copy()
df_pair = df_pair[df_pair['treatment'].isin(['Yes', 'No'])]  # Filter out 'Unknown' or unclear values

# Optional: label encode treatment for color grouping
df_pair['Treatment_Label'] = df_pair['treatment'].map({'Yes': 1, 'No': 0})

# Pairplot using numerical features
sns.pairplot(df_pair, vars=['Age'], hue='treatment', palette='Set1', diag_kind='kde')
plt.suptitle('Pair Plot: Age vs Treatment', y=1.02)
plt.show()
```

* To visually explore the relationship between age and mental health treatment. Helps detect age-related patterns in treatment-seeking behavior.
* Younger professionals (ages 20–35) are more likely to seek treatment, while older age groups (40+) report lower treatment rates.
* Enables age-specific targeting for mental health resources.
* Older employees may be underserved or face stigma — addressing this can improve inclusion, retention, and workplace wellbeing.

# 17. Mental Health Treatment by Country
